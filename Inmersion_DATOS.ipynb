```python
import matplotlib.pyplot as plt

promedio_area_por_barrio = inmuebles.groupby('Barrio')['Area'].mean()


top_10_barrios_por_area = promedio_area_por_barrio.nlargest(10)

ax = top_10_barrios_por_area.plot(kind='barh')

ax.set_title('Top 10 de Barrios con Mayor Promedio de Área')
ax.set_xlabel('Área Promedio')
ax.set_ylabel('Barrio')

for i, v in enumerate(top_10_barrios_por_area):
    ax.annotate("{:.2f}".format(v), xy=(v, i), va='center')

plt.show()




```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

tipos_incluidos = ['Edificio', 'Local', 'Finca']

plt.figure(figsize=(12,10))
grafica = sns.histplot(data=inmuebles[inmuebles['Tipo'].isin(tipos_incluidos)], x= 'Precio_Millon', kde=True, hue='Tipo')


grafica.set_title('Distribucion de Valores de los inmuebles en Bogota')
plt.xlim((50,4000))
plt.savefig('/content/drive/MyDrive/grafica_valor.png', format = 'png')
plt.show()
```

```python
import pandas as pd

# Agrupar por barrio y calcular el promedio de precios por metro cuadrado
precio_m2_por_barrio = inmuebles.groupby('Barrio')['Precio_Millon'].mean() / inmuebles.groupby('Barrio')['Area'].mean()

# Convertir a DataFrame y ordenar por precio por metro cuadrado
precio_m2_por_barrio = pd.DataFrame({'Precio por m2': precio_m2_por_barrio}).sort_values('Precio por m2', ascending=False)

# Mostrar los primeros 10 resultados
print(precio_m2_por_barrio)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculamos el precio por metro cuadrado para cada barrio
inmuebles_por_barrio = inmuebles.groupby('Barrio').mean()
inmuebles_por_barrio['Precio_m2'] = inmuebles_por_barrio['Precio_Millon'] / inmuebles_por_barrio['Area']
inmuebles_por_barrio.sort_values(by='Precio_m2', ascending=False, inplace=True)

# Creamos un gráfico de barras con Seaborn
plt.figure(figsize=(8,27))
grafica = sns.barplot(x='Precio_m2', y=inmuebles_por_barrio.index, data=inmuebles_por_barrio)

# Agregamos las etiquetas del precio por metro cuadrado en cada barra
for i, precio in enumerate(inmuebles_por_barrio['Precio_m2']):
    grafica.text(precio+0.5, i, '${:.2f}'.format(precio), va='center')

# Configuramos los títulos y etiquetas del gráfico
grafica.set_title('Precio por metro cuadrado en cada barrio')
grafica.set_xlabel('Precio por metro cuadrado (millones de pesos)')
grafica.set_ylabel('Barrio')

plt.tight_layout()
plt.show()

```
